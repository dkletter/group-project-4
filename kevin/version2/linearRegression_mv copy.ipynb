```python
# Encode categorical data to numerical
mv_encoded = pd.get_dummies(mv_drop1_df)
mv_encoded.shape
```

```python
# Split data into features and target
X = mv_encoded.drop(columns=['imdbrating'], axis=1)
y = mv_encoded['imdbrating']
```

```python
X[:1]
```

```python
mv_encoded.drop(columns=['imdbrating'], axis=1)
```

```python
# Split data into training and testing sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state= 42)
```

```python
X_test[:1]
```

```python
# Standardize the datset
from sklearn.preprocessing import StandardScaler

X_scaler = StandardScaler().fit(X_train)
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# set up linear regression
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(X_train_scaled, y_train)

training_score = model.score(X_train_scaled, y_train)
testing_score = model.score(X_test_scaled, y_test)
```
