```python
df = pd.read_csv('one_hot_encode.csv')
df.head(5)
```

```python
df.isnull().sum()
```

```python
df["rated"].fillna("NA", inplace = True)
df["director"].fillna("NA", inplace = True)
df["writer"].fillna("NA", inplace = True)
df["language"].fillna("NA", inplace = True)
df["country"].fillna("NA", inplace = True)
df["awards"].fillna("NA", inplace = True)
df["metascore"].fillna(0, inplace = True)
df["dvd"].fillna("NA", inplace = True)
df["boxoffice"].fillna(0, inplace = True)
df["budget"].fillna(0, inplace = True)
df["gross"].fillna(0, inplace = True)
# df.fillna(0, inplace=True)

df.isnull().sum()
```

```python
movie = df.drop(columns = ['title', 'released','runtime','writer','actors','plot','language','country','awards','ratings','metascore','imdbvotes','dvd','response','year','boxoffice','internet_movie_database_rating','rotten_tomato_rating','budget','gross','metacritic_rating'], axis=1)
movie.head(5)
```

```python
df2 = movie.drop(columns=['imdbrating'], axis=1)
df2.head(5)
```

```python
classonly = df2[['rated','director','genre','star']]
classonly.head(5)
```

```python
classonly[classonly['rated']=='PG-13']
```

```python
classonly.loc[classonly['rated'] == 'TV-MA', 'rated'] = 'R'
classonly.loc[classonly['rated'] == 'NC-17', 'rated'] = 'R'
classonly.loc[classonly['rated'] == 'TV-14', 'rated'] = 'PG-13'
classonly.loc[classonly['rated'] == 'TV-PG', 'rated'] = 'PG'
classonly.loc[classonly['rated'] == 'Not Rated', 'rated'] = 'NA'
classonly.loc[classonly['rated'] == 'Unrated', 'rated'] = 'NA'
```

```python
classonly['rated'].value_counts()
```

```python
stars = classonly.star.value_counts()
stars
```

```python
stars_filter = stars[stars<10]
stars_filter
```

```python
stars_replace = stars[stars<10].index.to_list()
for star in stars_replace:
  classonly['star'] =  classonly['star'].replace(star,'Others')

classonly['star'].value_counts()
```
